# Calculate basic statistics for the mean, median, mode, minimum, and maximum salary, determine the salary range, and find the standard deviation 

```python
TotalPay_stats = salaries['TotalPay'].describe()
print("إحصائيات الراتب:")
print(TotalPay_stats)
```

```python
median_salary = salaries['TotalPay'].median()
print("الوسيط:", median_salary)
```

```python
mode_salary = salaries['TotalPay'].mode()
print("المتوسط:")
print(mode_salary)
```

```python
min_salary = salaries['TotalPay'].min()
max_salary = salaries['TotalPay'].max()
print("الحد الأدنى للراتب:", min_salary)
print("الحد الأقصى للراتب:", max_salary)
```

```python
std_salaries = salaries['TotalPay'].std()
print("الانحراف المعياري للراتب:", std_salaries)
```
